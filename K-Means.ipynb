```python
sse = []
r1 = range(1,10)
for k in r1:
    km = KMeans(n_clusters=k)
    km.fit(x)
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(r1,sse)
```

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.8,min_samples=14)
db.fit(x)
```

```python
cluster=df[db.labels_!=-1]
outlier=df[db.labels_==-1]
colors= db.labels_
ccluster=colors[colors!=-1]
ocluster='black'
```

```python
plt.scatter(cluster['petal length (cm)'],cluster['petal width (cm)'],c=ccluster,edgecolors='black',s=50)
plt.scatter(outlier['petal length (cm)'],outlier['petal width (cm)'],c=ocluster,edgecolors='black',s=50)
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
```
